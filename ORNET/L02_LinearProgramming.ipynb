<h2>ปัญหากำหนดการเชิงเส้น</h2>
<ul>
<li><span style="text-decoration: underline;"><strong>ตัวแปรตัดสินใจ (Decision variables)</strong></span>: คือตัวแปรที่ต้องตัดสินหาค่าที่เหมาะที่สุดสำหรับฟังก์ชันจุดประสงค์ของปัญหากำหนดการเชิงเส้น ($x_j, j = 1, 2, &hellip;, n$)</li>
<li><span style="text-decoration: underline;"><strong>ฟังก์ชันจุดประสงค์ (Objective function)</strong></span>: เป็นฟังก์ชันที่ปัญหากำหนดการเชิงเส้นต้องการหาค่าสูงสุด/ต่ำสุดด้วยตัวแปรตัดสินใจ ($z = c_1 x_1 + c_2 x_2 + &hellip; + c_n x_n$) ตามเงื่อนไขบังคับที่กำหนด</li>
<li><span style="text-decoration: underline;"><strong>เงื่อนไขบังคับ (Constraints)</strong></span>: ซึ่งประกอบด้วยสมการหรืออสมการเชิงเส้นที่เกี่ยวข้องกับตัวแปรตัดสินใจ ($a_1 x_1 + a_2 x_2 + &hellip; + a_n x_n \leq b_1$)</li>
<li>การแปลงเงื่อนไขบังคับจาก อสมการเป็นสมการ หรือสมการเป็นอสมการทำได้โดย</li>
</ul>
<p>$$a_1 x_1 + a_2 x_2 + ... + a_n x_n \leq b \Rightarrow$$ $$a_1 x_1 + a_2 x_2 + ... + a_n x_n + w = b, w \geq 0$$</p>
<p>$$a_1 x_1 + a_2 x_2 + ... + a_n x_n = b \Rightarrow$$ $$a_1 x_1 + a_2 x_2 + ... + a_n x_n \leq b $$ $$a_1 x_1 + a_2 x_2 + ... + a_n x_n \geq b$$</p>
<ul>
<li>รูปแบบบัญญัติของปัญหากำหนดการเชิงเส้น (LP)</li>
</ul>
<p>$$\begin{array}{clcl} \mbox{(LP)} & \mbox{maximize} & c_1 x_1 + c_2 x_2 + ... + c_n x_n \\ & \mbox{subject to} & a_{1 1} x_1 + a_{1 2} x_2 + ... + a_{1 n} x_n \leq b_1 \\ & & a_{2 1} x_1 + a_{2 2} x_2 + ... + a_{2 n} x_n \leq b_2 \\ & & ... \\ & & a_{m 1} x_1 + a_{m 2} x_2 + ... + a_{m n} x_n \leq b_m \\ & & x_1, x_2, ..., x_n \geq 0 \end{array}$$</p>

<h2>ความเป็นไปได้ของผลเฉลยของปัญหากำหนดการเชิงเส้น</h2>
<ul>
<li>เซตของผลเฉลยไม่เป็นเซตว่างและมีผลเฉลยที่เหมาะที่สุด (Feasible and Optimal)</li>
<li>เซตของผลเฉลยเป็นเซตว่าง (Infeasible)</li>
<li>เซตของผลเฉลยไม่เป็นเซตว่างและไม่มีผลเฉลยที่เหมาะที่สุด หรือปัญหาไม่มีขอบเขต (Feasible and unbounded)</li>
</ul>
<h3>ตัวอย่าง</h3>
<p>$$\begin{array}{lcl} \mbox{maximize} & -x_1 + 3 x_2 - 3 x_3 \\ \mbox{subject to} & 3 x_1 - x_2 - 2 x_3 \leq 7 \\ & -2 x_1 - 4 x_2 + 4 x_3 \leq 3 \\ & x_1 - 2 x_3 \leq 4 \\ & -2 x_1 + 2 x_2 + x_3 \leq&nbsp; 8 \\ & 3 x_1 \leq 5 \\ & x_1, x_2, x_3 \geq 0 \end{array}$$</p>
<p>เขียนรูปใหม่พร้อมเพิ่มตัวแปรช่วย (slack variables)</p>
<p>$$\begin{array}{lcccl} \mbox{maximize} & z & = & -x_1 + 3 x_2 - 3 x_3 \\  \mbox{subject to} & w_1 & = & 7 - 3 x_1 + x_2 + 2 x_3 \\ & w_2 & = & 3 + 2 x_1 + 4  x_2 - 4 x_3 \\ & w_3 & = & 4 - x_1 + 2 x_3 \\ & w_4 & = & 8 + 2 x_1 - 2 x_2 -  x_3 \\ & w_5 & = & 5 - 3 x_1 \\ & & & x_1, x_2, x_3, w_1, w_2, w_3, w_4, w_5 \geq 0  \end{array}$$</p>
<ul>
<li>รูปแบบที่เขียนเรียกว่า dictionary</li>
<li>การกำหนดให้ $x_1, x_2, x_3$ เป็น 0 ทำให้เราสามารถอ่านค่าของตัวแปรใหม่ได้เป็น $$w_1 = 7, w_2 = 3, w_3 =4, w_4 = 8, w_5 = 5$$ This specific solution is เรียกผลเฉลยของ dictionary</li>
<li>ตัวแปรตาม (Dependent variables) ทางด้านซ้ายเรียกตัวแปรพื้นฐาน (basic variables)</li>
<li>ตัวแปรอิสระ (Independent variables) ทางด้านขวา เรียกตัวแปรไม่พื้นฐาน (nonbasic variables)</li>
<li>ตัวแปรทุกตัวมีค่าไม่เป็นลบ เรียกผลเฉลยที่ได้ว่าผลเฉลยที่เป็นไปได้ (feasible)</li>
<li>แน่นอนว่าผลเฉลยเริ่มต้นอาจไม่เป็นผลเฉลยที่เป็นไปได้</li>
</ul>

<h2>วิธีซิมเพล็กซ์</h2>
<ul>
<li>รอบแรก 
<ul>
<li>หาตัวแปรทางด้านซ้ายที่มีค่าสัมประสิทธิ์เป็นบวกในฟังก์ชันจุดประสงค์ที่เมื่อเพิ่มค่าของตัวแปรแล้วจะทำให้ค่าฟังก์ชันจุดประสงค์เพิ่มขึ้น (จากตัวอย่างคือตัวแปร $x_2$)</li>
<li>สำหรับตัวแปรพื้นฐานที่ตัวที่ได้รับผลกระทบจากตัวแปร $x_2$ เลือกพิจารณาเฉพาะตัวแปรพื้นฐานที่มีค่าเล็กลง</li>
<li>เลือกตัวแปรพื้นฐานตัวแรกที่ทำให้ค่าเป็นศูนย์ เมื่อเพิ่ม $x_2$ (จากตัวอย่างคือตัวแปร $w_4$)</li>
<li>ตัวแปร $x_2$ ต้องเป็นตัวแปรพื้นฐานแทนตัวแปร $w_4$ ซึ่งเป็นตัวแปรไม่พื้นฐาน</li>
<li>เราเรียกวิธีการดังกล่าวว่า การหมุน (pivot)</li>
<li>หลังการหมุนครั้งแรก เราได้ว่า</li>
</ul>
</li>
</ul>
<p>$$\begin{array}{lcccl} \mbox{maximize} & &zeta; & = & 12 + 2 x_1 - 1.5  w_4 - 4.5 x_3 \\  \mbox{subject to} & w_1 & = & 11 - 2 x_1 - 0.5 w_4 + 1.5 x_3 \\ & w_2 & = & 19 + 6 x_1 - 2 w_4 - 6 x_3 \\  & w_3 & = & 4 - x_1 + 2 x_3 \\ & x_2 & = & 4 + x_1 - 0.5 w_4 -  0.5 x_3 \\ & w_5 & = & 5 - 3 x_1 \\ & &  & x_1, x_2, x_3, w_1, w_2, w_3, w_4, w_5 \geq 0  \end{array}$$</p>
<ul>
<li>ต่อจากนั้น เราใช้หลักการเดียวกันเพื่อหาคู่ตัวแปรที่จะสลับระหว่างตัวแปรไม่พื้นฐานเป็นตัวแปรพื้นฐาน 
<ul>
<li>หลังจากหมุน แล้วได้ว่า</li>
</ul>
</li>
</ul>
<p>$$\begin{array}{lcccl} \mbox{maximize} & z & = & 46/3 - 2/3 w_5 -  1.5  w_4 - 4.5 x_3 \\  \mbox{subject to} & w_1 & = & 23/3 + 2/3  w_5 - 0.5 w_4 + 1.5 x_3 \\ & w_2 & = & 29 - 2 w_5 - 2 w_4 -  6 x_3 \\  & w_3 & = & 7/3 + 1/3 w_5 + 2 x_3 \\ & x_2 & =  & 17/3 - 1/3 w_5 - 0.5 w_4 -  0.5 x_3 \\ & x_1 & = & 5/3&nbsp; - 1/3 w_5  \\ & &  & x_1, x_2, x_3, w_1, w_2, w_3, w_4, w_5 \geq 0   \end{array}$$</p>
<ul>
<li>เราพบว่าการเลือกเป็นไปไม่ได้อีก แสดงว่าเราอยู่ ณ ผลเฉลยที่เหมาะที่สุดแล้ว</li>
</ul>


<h2>วิธีซิมเพล็กซ์ที่ผลเฉลยเริ่มต้นเป็นผลเฉลยที่เป็นไปได้</h2>
<ul>
<li>พิจารณาปัญหากำหนดการเชิงเส้นในรูปแบบบัญญัติ (Canonical form)</li>
</ul>
<p>$$\begin{array}{clcl} \mbox{(LP)} & \mbox{maximize} & \sum_{i = 1}^{n} c_i x_i & \\ & \mbox{subject to} & \sum_{j = 1}^{n}  a_{i j} x_j \leq b_i & i = 1, ..., m \\ &&nbsp; & x_j \geq 0 & \end{array}$$</p>
<ul>
<li>เราเริ่มจากการกำหนดตัวแปรพื้นฐานและตัวแปรไม่พื้นฐานให้เรียกว่า $x$ กล่าวคือ $(x_1, ..., x_n, w_1, ..., w_m)$ = $(x_1, ..., x_n, x_{n + 1}, ..., x_{n + m})$</li>
<li>เริ่มต้นได้ว่า</li>
</ul>
<p>$$\begin{array}{ll} z = \sum_{i = 1}^{n} c_i x_i & \\ x_{n+i} = b_i - \sum_{j = 1}^{n} a_{i j} x_j & i = 1, 2, ..., m \end{array}$$</p>
<ul>
<li>สำหรับเซตของดัชนี $S$={$1, 2, &hellip;, n+m$}, ให้ $B \subseteq S$ แทนเซตของดัชนีของตัวแปรพื้นฐานและให้ $N \subseteq S$ แทนเซตของดัชนีของตัวแปรไม่พื้นฐาน เมื่อเริ่มต้นเราได้ว่า $$B = \{n + 1, n + 2, ..., n + m\}, N = \{1, 2, ..., n\}$$</li>
<li>หลังการการหมุนผ่านไป เราอาจหยุดอยู่ที่</li>
</ul>
<p>$$\begin{array}{ll} z = \bar{z} \sum_{i = 1}^{n} \bar{c}_i x_i & \\ x_{i} =  \bar{b}_i - \sum_{j = 1}^{n} \bar{a}_{i j} x_j & i = 1, 2, ..., m \end{array}$$</p>
<ul>
<li><span style="text-decoration: underline;"><strong>ตัวแปรเข้า (Entering variable)</strong></span>: คือตัวแปรที่เปลี่ยนจากตัวแปรไม่พื้นฐานเป็นตัวแปรพื้นฐาน สำหรับแนวคิดของ Dantzig เลือกตัวแปรเข้าที่มีค่าสัมประสิทธิ์ในฟังก์ชันจุดประสงค์ปัจจุบันที่มากที่สุด $$\{x_j| j \in N, c_j \mbox{ มีค่าสูงสุด}\}$$</li>
<li><span style="text-decoration: underline;"><strong>ตัวแปรออก (Leaving variable)</strong></span>: คือตัวแปรที่เปลี่ยนจากตัวแปรพื้นฐานเป็นตัวแปรไม่พื้นฐาน สมมติว่าตัวแปร $x_k$ เป็นตัวแปรเข้า ตัวแปรออกถูกเลือกโดยที่เป็นตัวแปรแรกสุดที่เมื่อค่าของตัวแปร $x_k$ เพิ่มขึ้นแล้วทำให้ค่าตัวแปรดังกล่าวเป็นศูนย์ $$\{x_i | i \in N, \frac{\bar{a}_{i j}}{\bar{b}_j} \mbox{ เป็นค่าสูงสุด}\}$$</li>
<li>หลักเกณฑ์การหมุน:กำหนดการเลือกตัวแปรเข้าและตัวแปรออกที่เหมาะสม</li>
</ul>


<h2>ปัญหาไม่มีขอบเขต</h2>
<ul>
<li>พิจารณาตัวแบบปัญหากำหนดการเชิงเส้น</li>
</ul>
<p>$$\begin{array}{ccl} z & = & \phantom{10} + 2 x_1 - x_2 + x_3 \\ w_1 & = &&nbsp; 4 + 5 x_1 - 3 x_2 + x_3 \\ w_2 & = & 10 +&nbsp; x_1 + 5 x_2 - 2 x_3 \\ w_3 & =&&nbsp; 7 \phantom{+ 10 x_1} + 4 x_2 - 3 x_3 \\ w_4 & = &&nbsp; 6 + 2 x_1 + 2 x_2 - 4 x_3 \\ w_5 & = &&nbsp; 6 + 3 x_1 + 3 x_3 \end{array}$$</p>
<ul>
<li>การเพิ่มขึ้นของตัวแปร $x_1$ หรือตัวแปร $x_3$ ทำให้ค่าฟังก์ชันจุดประสงค์เพิ่มขึ้น</li>
<li>แต่การเพิ่มของตัวแปร $x_1$ ไม่ส่งผลให้ตัวแปรพื้นฐานอื่น ๆ ลดลง</li>
<li>ดังนั้นการเพิ่มขึ้นของตัวแปร $x_1$ สามารถทำให้ค่าฟังก์ชันจุดประสงค์เพิ่มขึ้นโดยไม่มีขอบเขต</li>
<li>นี้คือวิธีการตรวจสอบว่าปัญหาที่กำหนดให้เป็นปัญหาที่มีขอบเขตหรือไม่สำหรับวิธีซิมเพล็กซ์</li>
</ul>

<h2>การตรวจสอบการไม่มีผลเฉลยของวิธีซิมเพล็กซ์</h2>
<ul>
<li>พิจารณาปัญหากำหนดการเชิงเส้น</li>
</ul>
<p>$$\begin{array}{lcl} \mbox{Maximize} & -3 x_1 + 4 x_2 &  \\  \mbox{subject to} & -4 x_1 - 2 x_2 \leq -8 & \\ & -2&nbsp; x_1 \leq -2 \\ & 3 x_1 + 2 x_2 \leq 10 \\ & - x_1 + 3 x_2 \leq 1 \\ & - 3 x_2 \leq -2 \\ & x_1,  x_2 \geq 0  \end{array}$$</p>
<ul>
<li>ปัญหาเฟส I:ปรับปัญหากำหนดการเชิงเส้นโดยการลบตัวแปรใหม่ $x_0$ ในทุกเงื่อนไขบังคับ แล้วพิจารณาฟังก์ชันจุดประสงค์เป็น $&ndash;x_0$</li>
</ul>
<p>$$\begin{array}{lcl} \mbox{Maximize} & x_0 &  \\   \mbox{subject to} & -x_0 - 4 x_1 - 2 x_2 \leq -8 & \\ & -x_0 - 2&nbsp; x_1  \leq -2 \\ & -x_0 + 3 x_1 + 2 x_2 \leq 10 \\ & -x_0 - x_1 + 3 x_2 \leq 1 \\  & -x_0 - 3 x_2 \leq -2 \\ & x_0, x_1,  x_2 \geq 0  \end{array}$$</p>
<ul>
<li>แน่นอนว่าถ้าค่าของตัวแปร $x_0$ ใหญ่พอ เราจะได้ว่า $x_1$ = 0 และ $x_2$ = 0</li>
<li>ในกรณีที่วิธีซิมเพล็กซ์จบลงที่ค่าฟังก์ชันจุดประสงค์ใหม่เป็นศูนย์ เราได้ว่าปัญหาเริ่มต้นเป็นปัญหาที่มีผลเฉลยที่เป็นไปได้</li>
<li>แน่นอนว่าเซตของตัวแปรพื้นฐานในปัญหาเฟส I สามารถใช้ในการเริ่มแก้ปัญหากำหนดการเชิงเส้นนี้ได้ เรียกปัญหาเฟส II โดยเราไม่สนใจค่าของตัวแปร $x_0$ อีกต่อไป</li>
<li>ในกรณีที่วิธีซิมเพล็กซ์จบลงที่ค่าฟังก์ชันจุดประสงค์น้อยกว่าศูนย์ เรากล่าวว่าปัญหาเริ่มต้นเป็นปัญหาที่ไม่มีผลเฉลย</li>
</ul>

<h2 style="font-size: 1.5em; text-align: left;">ขั้นตอนวิธีซิมเพล็กซ์ในรูปแบบเวกเตอร์และเมทริกซ์</h2>
<p>พิจารณาปัญหากำหนดการเชิงเส้นที่มีจุดกำเนิดเป็นหนึ่งในผลเฉลยที่เป็นไปได้</p>
<p>$$\begin{array}{lcl} \mbox{Maximize} &amp; {\bf c}^T {\bf x} &amp;  \\    \mbox{subject to} &amp; {\bf A x} \leq {\bf b} &amp; \\ &amp; {\bf x} \geq {\bf 0} &amp; \end{array}$$</p>
<p>ขั้นเริ่มต้น:กำหนดเซตของตัวแปรพื้นฐาน $IndB$ และเซตของตัวแปรไม่พื้นฐาน $IndN$</p>
<p>ทำให้เราแบ่ง</p>
<ul>
<li>เมทริกซ์ $\bf A$ ออกเป็นเมทริกซ์พื้นฐาน $\bf B$ ที่มีอินเวอร์ส และเมทริกซ์ไม่พื้นฐาน $\bf N$</li>
<li>ตัวแปร $\bf x$ ออกเป็นตัวแปรพื้นฐาน $\bf x_B$ กับตัวแปรไม่พื้นฐาน $\bf x_N$</li>
<li>สัมประสิทธิ์ของฟังก์ชันจุดประสงค์ $\bf c$ ออกเป็น $\bf c_B$ กับ $\bf c_N$</li>
</ul>
<p>พิจารณาระบบสมการเชิงเส้นที่เกิดจากเงื่อนไขบังคับของปัญหา $\bf A x$ = $\bf b$</p>
<p>$$\bf [B | N] \left[\begin{array}{c} \bf x_B \\ \bf x_N \end{array}\right] = b$$ เนื่องจาก $\bf B$ มีอินเวอร์ส ได้ว่า</p>
<p>$$\bf B x_B = b - N x_N$$ $$\bf x_B = B^{-1} b - B^{-1} N x_N$$</p>
<p>ในกรณีที่ $\bf x_N$ = $\bf 0$ เราได้ว่า $\bf x_B = B^{-1} b$</p>

```python
# Setting the problem in a standard form
m = 3
n = 2
c = [3, 2, 0, 0, 0]
A = [[-1, 3, 1, 0, 0], [1, 1, 0, 1, 0], [2, -1, 0, 0, 1]]
b = [12, 8, 10]

print("The problem is to maximize", "+".join(["%s x%s"%(c[i],i+1) for i in range(n)]))
print("subject to, ")
for k in range(m):
    print("+".join(["%s x%s"%(A[k][i],i+1) for i in range(n)]), "<=", b[k])
print(",".join(["x%s"%(i+1) for i in range(n)]), ">= 0")
```

```python
def simplex(c, A, b, m, n, showStep=True):
    """The simplex method:using the slack variables as basis.
       Written by Krung Sinapiromsaran
    """ 
    # Set the initial BFS
    IndB = [i+n for i in range(m)]
    IndN = [i for i in range(n)]
    if showStep:
        print("The basic and nonbasic index are ", IndB, IndN)
        
    while True:
       # Step 1: Solve Bx = b
       B = [[A[i][IndB[j]] for j in range(m)] for i in range(m)]
       BM = matrix(B)
       xB = BM.solve_right(vector(b))
       z = sum([c[IndB[i]]*xB[i] for i in range(m)])
       if showStep:
           print("The current BFS is ", B)
           print("xB is ", xB, " and current objective is ", z)

       # Step 2: Solve w^T B = c_B^T
       cB = [c[IndB[i]] for i in range(m)]
       w = BM.solve_left(vector(cB))
       if showStep:
           print("The current w is ", w)

       # Step 3: Determine the minimum z_k - c_k
       k0 = 0
       k = IndN[0]
       Ak = [A[i][k] for i in range(m)]
       minZC = sum([w[i]*Ak[i] for i in range(m)]) - c[k]
       for i in range(1,n):
           k = IndN[i]
           Ak = [A[j][k] for j in range(m)]
           if minZC > sum([w[j]*Ak[j] for j in range(m)]) - c[k]:
               k0 = i
               minZC = sum([w[j]*Ak[j] for j in range(m)]) - c[k]
       if showStep:
           print("The current in-variable is ", IndN[k0])

       # Step 4: Check for the optimality 
       if minZC >= 0:
           print("The optimal solution is found at ", xB, "with obj.=", z)
           return [xB, z, IndB] 

       # Step 5: Solve By = Ak
       k = IndN[k0]
       Ak = [A[i][k] for i in range(m)]
       yk = BM.solve_right(vector(Ak))

       # Step 6: Determine the unboundedness
       UnboundFlag = True
       for i in range(m):
           if yk[i] > 0:
               UnboundFlag = False
               break
       if UnboundFlag:
           print("The problem is unbounded.")
           return [xB, oo, IndB]

       # Step 7:Minimum ratio test to determine the out-variable
       r = -1
       minRatio = oo
       for i in range(m):
           if yk[i] > 0:
               if xB[i]/yk[i] < minRatio:
                   r = i
                   minRatio = xB[i]/yk[i]
       if showStep:
           print("The current out-variable is ", IndB[r])

       # Step 8:Swap basic and nonbasic
       if showStep:
           print("Swap between nonbasic:", k, " with a basic:", IndB[r])
       IndN[k0] = IndB[r]
       IndB[r] = k
```

```python
result = simplex(c, A, b, m, n, showStep=True)
```

```python
def simplexPath(c, A, b, m, n, showStep=True):
    """The simplex method:using the slack variables as basis.
       Written by Krung Sinapiromsaran
    """    
    Path = []
    SimplexPoint = [0 for i in range(n)]
    
    # Set the initial BFS
    IndB = [i+n for i in range(m)]
    IndN = [i for i in range(n)]
    if showStep:
        print("The basic and nonbasic index are ", IndB, IndN)
        
    while True:
       # Step 1: Solve Bx = b
       B = [[A[i][IndB[j]] for j in range(m)] for i in range(m)]
       BM = matrix(B)
       xB = BM.solve_right(vector(b))
       z = sum([c[IndB[i]]*xB[i] for i in range(m)])
       if showStep:
           print("The current BFS is ", B)
           print("xB is ", xB, " and current objective is ", z)

       for i in range(m):
           if IndB[i] < n:
               SimplexPoint[IndB[i]] = xB[i]
       Path.append(copy(SimplexPoint))

       # Step 2: Solve w^T B = c_B^T
       cB = [c[IndB[i]] for i in range(m)]
       w = BM.solve_left(vector(cB))
       if showStep:
           print("The current w is ", w)

       # Step 3: Determine the minimum z_k - c_k
       k0 = 0
       k = IndN[0]
       Ak = [A[i][k] for i in range(m)]
       minZC = sum([w[i]*Ak[i] for i in range(m)]) - c[k]
       for i in range(1,n):
           k = IndN[i]
           Ak = [A[j][k] for j in range(m)]
           if minZC > sum([w[j]*Ak[j] for j in range(m)]) - c[k]:
               k0 = i
               minZC = sum([w[j]*Ak[j] for j in range(m)]) - c[k]
       if showStep:
           print("The current in-variable is ", IndN[k0])

       # Step 4: Check for the optimality 
       if minZC >= 0:
           print("The optimal solution is found at ", xB)
           print(" and its objective is ", z)
           return [xB, z, IndB, Path] 

       # Step 5: Solve By = Ak
       k = IndN[k0]
       Ak = [A[i][k] for i in range(m)]
       yk = BM.solve_right(vector(Ak))

       # Step 6: Determine the unboundedness
       UnboundFlag = True
       for i in range(m):
           if yk[i] > 0:
               UnboundFlag = False
               break
       if UnboundFlag:
           print("The problem is unbounded.")
           return [xB, oo, IndB, Path]

       # Step 7:Minimum ratio test to determine the out-variable
       r = -1
       minRatio = oo
       for i in range(m):
           if yk[i] > 0:
               if xB[i]/yk[i] < minRatio:
                   r = i
                   minRatio = xB[i]/yk[i]
       if showStep:
           print("The current out-variable is ", IndB[r])

       # Step 8:Swap basic and nonbasic
       if showStep:
           print("Swap between nonbasic:", k, " with a basic:", IndB[r])
       IndN[k0] = IndB[r]
       IndB[r] = k
       if IndN[k0] < n:
           SimplexPoint[IndN[k0]] = 0
```

```python
resultPath = simplexPath(c, A, b, m, n, showStep=True)
```

<h2>การวิเคราะห์กรณีที่แย่ที่สุด</h2>
<ul>
<li>ใช้จำนวนการทำซ้ำเป็นเลขชี้กำลัง</li>
<li>ปัญหาของ Klee-Minty แสดงให้เห็นว่าหลักเกณฑ์การหมุนของ Dantzig ที่กำหนดโดยเลือกค่าสัมประสิทธิ์ที่ใหญ่ที่สุดอาจใช้จำนวนการหมุน $2^n-1$ เพื่อแก้ปัญหาที่มี $n$ ตัวแปรและ $n$ เงื่อนไขบังคับ กล่าวคือการเยี่ยมทุกจุดยอด $2^n$ ของลูกบาศก์ที่บิดไป</li>
