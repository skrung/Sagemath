<h2>2. Mathematical definition</h2>
<p>Given a finite list of real numbers of $n$ elements, $x_1, x_2, x_3, \ldots, x_n$. The ascending list of these numbers is defined as a permutation of the indices from {1, 2, 3, ..., $n$}, called $a_1, a_2, a_3, \ldots, a_n$, such that \[x_{a_1} \leq x_{a_2} \leq \cdots \leq x_{a_n}.\]</p>
<p>Note that $x_{a_1}$ is the minimum element and $x_{a_n}$ is the maximum element.<br /><br />The definition of the descending list is defined similarly.&nbsp; It is a permutation of {1, 2, 3, ..., $n$}, called $d_1, d_2, d_3, \ldots, d_n$, such that \[x_{d_1} \geq x_{d_2} \geq \cdots \geq x_{d_n}.\]</p>
<p>Note that $x_{d_1}$ is the maximum element and $x_{d_n}$ is the minimum element.<br /><br />The permutation space of these indices {1, 2, 3, ..., $n$}, $\Omega_n$, is defined as the set of all possible arrangements of 1, 2, 3, ..., $n$. That means $\Omega_n = \lbrace (k_1, k_2, k_3, ..., k_n) | k_1, k_2, k_3, \ldots, k_n \in \lbrace 1, 2, 3, \ldots, n \rbrace$ and all $k_i$'s are distinct $\rbrace$.</p>
<p>The <em>successive difference</em> of an element in the permutation space is defined as a function $d:\Omega_n \rightarrow \mathbb{R}^{+}$&nbsp; for $k_1, k_2, k_3, \ldots, k_n$ of {1, 2, 3, ..., $n$}<br />\[ d(k_1, k_2, \ldots, k_n) = \sum_{i = 1}^{n-1} \lvert x_{k_{i+1}} - x_{k_i} \rvert\]<br />where $\mathbb{R}^{+}$ is the set of nonnegative numbers.</p>

<p>For example, the initial list of four numbers (4, 3, 1, 7) has the successive difference<br />\[ d(1, 2, 3, 4) = |4 - 3|+|3 - 1|+|1 - 7| = 1 + 2 + 6 = 9.\]<br />Its ascending permutation is 3, 2, 1, 4 that is a sequence (1, 3, 4, 7) which has the successive difference<br />\[ d(3, 2, 1, 4) = |1 - 3| +|3 - 4|+|4 - 7| = 2 + 1 + 3 = 6. \]<br />Note that its descending permutation is the reverse of the above permutation, 4, 1, 2, 3, that is a sequence (7, 4, 3, 1) which has the successive difference<br />\[ d(4, 1, 2, 3) = |7 - 4|+|4 - 3|+|3 - 1| = 3 + 1 + 2 = 6.\]<br />It is clear that other permutations will have the successive difference bigger than 6.</p>

<h2>3. Successive difference</h2>
<p>I first show that the successive difference of the sorted list is the different of the maximum and the minimum of the list. The second theorem shows that any unsorted list of distinct elements has the successive difference bigger than the different between the maximum and the minimum. This implies that the sorted list will have the minimum successive difference.</p>

<p><strong>Theorem 1</strong>. The ascending sequence of the list of $n$ numbers has the minimum successive difference equal to the maximum subtracting the minimum.<br /><strong>Proof</strong>. Let $x_{a_1} \leq x_{a_2} \leq \cdots \leq x_{a_n}$&nbsp; be the ascending sequence of the list.<br />\[\begin{array}{lcl} d(a_1, a_2, &hellip;, a_n) & = & \displaystyle \sum_{i = 1}^{n - 1} |x_{a_{i+1}} - x_{a_i}| \\</p>
<p>& = & \displaystyle |x_{a_2} - x_{a_1}| + |x_{a_3} - x_{a_2}| + \cdots + |x_{a_n} - x_{a_{n - 1}}| \\</p>
<p>& = & \displaystyle (x_{a_2} - x_{a_1}) + (x_{a_3} - x_{a_2}) + \cdots + (x_{a_n} - x_{a_{n - 1}}) \\</p>
<p>& = & \displaystyle x_{a_n} - x_{a_1}</p>
<p>\end{array}\]</p>
<p><br />Therefore, the successive difference is equal to the maximum ($x_{a_n}$) minus the minimum ($x_{a_1}$) as state. It is easy to see that the descending sequence has the same successive difference,&nbsp; only switch the sign of each term.</p>

<p><strong>Corollary 1. </strong>The descending sequence of the list has the successive difference equal to the maximum subtracting the minimum.<br /><br />Note that $x_{d_1} \geq x_{d_2} \geq \cdots \geq x_{d_n}$ is equivalent to $- x_{d_1} \leq - x_{d_2} \leq \cdots \leq - x_{d_n}$. Apply theorem 1 to this list, then the successive difference is $-x_{d_n} - (- x_{d_1}) = x_{d_1} - x_{d_n}$ which is the maximum minus the minimum.</p>

<p><strong>Theorem 2</strong>. Let $S = (x_1, x_2, &hellip;, x_n)$ be the list of all distinct numbers that has $l$ as the minimum and $u$ as the maximum. If $(x_1, x_2, &hellip;, x_n)$ is not in the ascending nor descending order then $d(1, 2, 3, &hellip;, n) &gt; u - l$.<br /><strong>Proof.</strong> I will use the contradiction proof.</p>
<p>Let $S = (x_1, x_2, &hellip;, x_n)$ be the list of all distinct numbers that  has $l$ as the minimum and $u$ as the maximum. Moreover, $(x_1, x_2, &hellip;, x_n)$  is not sorted in the ascending nor descending order.</p>
<p>Identify the index of $l$ and $u$, said $x_i = l$ and $x_j = u$.</p>
<p>If $i \lt j$ then we go through the list $(x_i, x_{i+1}, ..., x_{j})$. If it is sorted then $d(i, i+1, i+2, ..., j) = u - l$.</p>
<p>However, the original list is not sorted implies that $i \neq 1$ or $j \neq n$. Therefore, we have at least one more nonnegative value to add to the total successive difference, i.e., $d(1, 2, 3, &hellip;, n) \gt u - l$.</p>
<p>If it is not sorted then we scan only for subset that is sorted, then $d(i, ..., j) = u - l$. When we put back the unsorted element, we only increment this number. Hence, $d(1, 2, 3, ..., n) &gt; u - l$.</p>
<p>If $i \gt j$ then the similar proof can be adapted.</p>

<h3>4.1 Selection sort</h3>
<p>Selection sort is one of the studied sorting algorithm that uses the greedy strategy. The sort is performed iteratively from 1 to $n-1$ where $n$ is the length of the input array. The array is considered as two subarrays, the sorted subarray and the unknown subarray. In each iteration, the greedy strategy picks the smallest element from the unknown subarray and swap it to the front of the subarray. Here is my simple pseudocode for selection sort algorithm.</p>

<p><span style="font-family: courier new,courier;">SELECTIONSORT(A, n)<br />INPUT: A is an array of length n say A(1:n)<br />OUTPUT: A is sorted in place.<br />1. for i = 1 to n-1<br />2.&nbsp;&nbsp;&nbsp;&nbsp; minx = A(i); k = i;<br />3.&nbsp;&nbsp;&nbsp;&nbsp; for j = i+1 to n<br />4.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if(A(j) &lt; minx) then<br />5.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; minx = A(j); k = j;<br />6.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endif<br />7.&nbsp;&nbsp;&nbsp;&nbsp; endfor<br />8.&nbsp;&nbsp;&nbsp;&nbsp; A(k) = A(i); A(i) = minx;<br />9. endfor</span></p>


```python
L = [4, 3, 1, 7]
def selectionsort(A):
    n = len(A)
    for i in range(n-1):
        minx = A[i]
        k = i
        for j in range(i+1, n):
            if A[j] < minx:
                minx = A[j]
                k = j
        A[k] = A[i]
        A[i] = minx
    return A
        
print('The original list is ', L)
print('The sort list is ', selectionsort(L))
```

<p>To use the successive difference, I have to compute the successive difference of sorting list which is equal to maximum - minimum of A, called this sort_sd. I also compute the original successive difference of A kept in succ. During the selection of minimum, I then update this succ. If the value of succ is less than or equal to sort_sd, then the list is sorted.</p>

<p><span style="font-family: courier new,courier;">SELECTION_SDSORT(A, n)<br />INPUT: A is an array of length n say A(1:n)<br />OUTPUT: A is sorted in place.<br />1. sort_sd = max(A) - min(A)<br />2. succ = successive(A, n)<br />3. for i = 1 to n-1<br />4.&nbsp;&nbsp;&nbsp; if(succ &lt;= sort_sd) then return<br />5.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; minx = A(i); k = i;<br />6.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; for j = i+1 to n<br />7.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if(A(j) &lt; minx) then<br />8.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; minx = A(j); k = j;<br />9.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endif<br />10.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endfor<br />11.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; before = fabs(A(i+1)-A(i))+A(k-1);<br />12.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; A(k) = A(i); A(i) = minx;<br />13.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; after&nbsp; = A(i+1)+fabs(A(k)-A(k-1));<br />14.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if(k &lt; n-1) {before +=A(k+1)-minx;<br />15.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; after += fabs(A(k+1)-A(k));}<br />16.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if(i &gt; 0){before +=A(k); after +=minx;}<br />17.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; succ += after - before;<br />18. endfor</span></p>


```python
L = [4, 3, 1, 7]
def selectionsdsort(A):
    n = len(A)
    sort_sd = max(A) - min(A)
    succ = sum([abs(A[i]-A[i+1]) for i in range(n-1)])
    for i in range(n-1):
        if succ <= sort_sd:
            return A
        minx = A[i]
        k = i
        for j in range(i+1, n):
            if A[j] < minx:
                minx = A[j]
                k = j
        A[k] = A[i]
        A[i] = minx
        before = abs(A[i+1]-A[i])+A[k-1]
        after = A[i+1] + abs(A[k]-A[k-1])
        if k < n-1:
            before = before + A[k+1]-minx
            after  = after + abs(A[k+1]-A[k])
        if i > 0:
            before = before + A[k]
            after  = after + minx
        succ = succ + after - before
    return A
        
print('The original list is ', L)
print('The sort list is ', selectionsdsort(L))
```

<p>As the pseudocode shown above, the succ is updated on line 17 where after and before updates on line 11 and 13-16. The time spent is constant with respect to the selection sort algorithm. I will compare the total running time of the regular selection sort, the selection sort that performs checking of the unsorted subarray and the selection sort that uses the successive difference on the simulation runs.</p>

<h3>4.2 Insertion sort</h3>
<p>Insertion sort is an internal sorting the perform the best running time of O($n$). It perform iteratively from 2 to $n$ by locating the place to put the next element within the known sorted list in the front. Normally, if the list is already sorted, the insertion sort will go through element one by one and stop.</p>

<p><span style="font-family: courier new,courier;">INSERTIONSORT(A, n)<br />INPUT: A is an array of length n say A(1:n)<br />OUTPUT: A is sorted in place.<br /> 1.&nbsp;for i = 2 to n<br /> 2.&nbsp;&nbsp;&nbsp;&nbsp; value = A(i); j = i - 1;<br /> 3.&nbsp;&nbsp;&nbsp;&nbsp; done = false<br /> 4.&nbsp;&nbsp;&nbsp;&nbsp; repeat<br /> 5.&nbsp;&nbsp;&nbsp;&nbsp; if A(j) &gt; value then<br /> 6.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; A(j+1) = A(j)<br /> 7.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; j = j - 1<br /> 8.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if j &lt; 0 then<br /> 9.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; done = true<br /> 10.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endif<br /> 11.&nbsp;&nbsp;&nbsp; else<br /> 12.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; done = true<br /> 13.&nbsp;&nbsp;&nbsp; endif<br /> 14.&nbsp;&nbsp;&nbsp; until done<br /> 15.&nbsp;&nbsp;&nbsp; A(j+1) = value<br /> 16.&nbsp;endfor</span></p>

```python
L = [4, 3, 1, 7]
def insertsort(A):
    n = len(A)
    for i in range(1, n):
        value = A[i]
        j = i - 1
        while True:
           if A[j] > value:
               A[j+1] = A[j]
               j = j - 1
               if j < 0:
                   break
           else:
               break
        A[j+1] = value
    return A
        
print('The original list is ', L)
print('The sort list is ', insertsort(L))
```

<p>To use the successive difference, I have to compute the successive  difference of sorting list which is equal to maximum - minimum of A,  called this sort_sd. I also compute the original successive difference  of A kept in succ. During the insertion of minimum, I then update this  succ. If the value of succ is less than or equal to sort_sd, then the  list is sorted.</p>

<p><span style="font-family: courier new,courier;">INSERTION_SDSORT(A, n)<br />INPUT: A is an array of length n say A(1:n)<br />OUTPUT: A is sorted in place.<br /> 1. sort_sd = max(A) - min(A)<br /> 2. succ = sucList(A, n)<br /> 3. for i = 2 to n<br /> 4.&nbsp;&nbsp;&nbsp; if(succ[1] &lt;= sort_sd) then return<br /> 5.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; value = A(i); j = i - 1;<br /> 6.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; done = false<br /> 7.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; repeat<br /> 8.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if A(j) &gt; value then<br /> 9.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; A(j+1) = A(j)<br /> 10.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; j = j - 1<br /> 11.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if j &lt; 0 then<br /> 12.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; done = true<br /> 13.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endif<br /> 14.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; else<br /> 15.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; done = true<br /> 16.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; endif<br /> 17.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; until done<br /> 18.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; succ[1] = (A[i]-A[1])+abs(A[i+1]-A[i])+succ[i+1]<br /> 19.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; A(j+1) = value<br /> 20.&nbsp;endfor </span></p>


```python
L = [4, 3, 1, 7]
def succ(A):
    SA = copy(A)
    minvalue = A[0]
    maxvalue = A[0]
    for i in range(1, len(A)):
        SA[i-1] = abs(A[i-1]-A[i])
        if minvalue > A[i]:
            minvalue = A[i]
        elif maxvalue < A[i]:
            maxvalue = A[i]
    SA[len(SA)-1] = 0
    for i in range(len(SA)-2):
        SA[len(SA)-i-2] = SA[len(SA)-i-1]+SA[len(SA)-i-2]
    SA[0] = SA[0]+SA[1]
    SA[len(A)-1] = maxvalue-minvalue
    return SA
    
def insertsdsort(A):
    n = len(A)
    SA = succ(A)
    sort_sd = SA[len(SA)-1]
    SA[len(SA)-1] = 0
    for i in range(1, n):
        if SA[0] <= sort_sd:
            return A
        value = A[i]
        j = i - 1
        while True:
           if A[j] > value:
               A[j+1] = A[j]
               j = j - 1
               if j < 0:
                   break
           else:
               break
        A[j+1] = value
        if i < n-1:
           SA[0] = (A[i]-A[0])+abs(A[i+1]-A[i])+SA[i+1]
    return A
        
print('The original list is ', L)
print('The sort list is ', insertsdsort(L))
```

<p>The pseudocode for updating the successive difference for the insertion sort in line 18 uses the property that the front list before $i$ is sorted and the successive difference computes from the tail elements is performed.</p>

```python
%%cython
def selectionsort(A):
    n = len(A)
    for i in range(n-1):
        minx = A[i]
        k = i
        for j in range(i+1, n):
            if A[j] < minx:
                minx = A[j]
                k = j
        A[k] = A[i]
        A[i] = minx
    return A
def selectionsdsort(A):
    n = len(A)
    sort_sd = max(A) - min(A)
    succ = sum([abs(A[i]-A[i+1]) for i in range(n-1)])
    for i in range(n-1):
        if succ <= sort_sd:
            return A
        minx = A[i]
        k = i
        for j in range(i+1, n):
            if A[j] < minx:
                minx = A[j]
                k = j
        A[k] = A[i]
        A[i] = minx
        before = abs(A[i+1]-A[i])+A[k-1]
        after = A[i+1] + abs(A[k]-A[k-1])
        if k < n-1:
            before = before + A[k+1]-minx
            after  = after + abs(A[k+1]-A[k])
        if i > 0:
            before = before + A[k]
            after  = after + minx
        succ = succ + after - before
    return A
def selectionchsort(A):
    n = len(A)
    for i in range(n-1):
        succ = sum([abs(A[j]-A[j+1]) for j in range(n-1)])
        if succ == max(A) - min(A):
            return A
        minx = A[i]
        k = i
        for j in range(i+1, n):
            if A[j] < minx:
                minx = A[j]
                k = j
        A[k] = A[i]
        A[i] = minx
    return A
def instSel(filename, numInstances, choice):
    infile = open(filename, "r")
    for k in range(numInstances):
        line = infile.readline()
        seq = line.split(',')
        seq = [int(seq[i]) for i in range(len(seq))]
        if choice == 0:
            selectionsort(seq)
        if choice == 1:
            selectionsdsort(seq)
        if choice == 2:
            selectionchsort(seq)
    infile.close()
```
